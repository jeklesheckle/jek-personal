```python
import requests as rq
import pandas as pd
import json
import urllib
import time
```

```python
schema_response = rq.get("https://api.opendota.com/api/schema")
schema_response_obj = json.loads(schema_response.content)
```

```python
# table_name, column_name, data_type
table_names = []
for entry in schema_response_obj:
    print(entry["table_name"] + "\t\t" + entry["column_name"] + "\t\t" + entry["data_type"])
    table_names.append(entry["table_name"])
```

```python
def make_request(player_id = None):
    custom_string = player_id
    if not player_id:
        custom_string = input("input player ID: ")
    req_percent_encoded = urllib.parse.quote(custom_string.encode("utf8"))
    BASE_REQUEST_STRING = "https://api.opendota.com/api/players/"
    request_string = BASE_REQUEST_STRING + req_percent_encoded
    # print(request_string)
    response = rq.get(request_string)
    return response
```

```python
def get_medal_from_rank_tier(rank_tier):
    medals = {1: "Herald", 2: "Guardian", 3: "Crusader", 4: "Archon", 5: "Legend", 6: "Ancient", 7: "Divine", 8: "Immortal"}

    rank_tier = int(rank_tier)
    if rank_tier == 0:
        return "Unranked"
    pips = rank_tier % 10
    medal = rank_tier // 10
    print("medal: " + str(medal) + " pips: " + str(pips))

    return medals[medal] + " " + str(pips)

def get_rank_tier(player_id):
    if type(player_id) is int:
        player_id = str(player_id)
    response = make_request(player_id)
    player_obj = json.loads(response.content)
    # print(player_obj)
    rank_tier = player_obj["rank_tier"]
    if rank_tier is None:
        return 0
    else:
        return rank_tier
```
